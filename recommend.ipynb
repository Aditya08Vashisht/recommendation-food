```python
# Load your food dataset
df = pd.read_csv('C:\\Users\\aditya vashisht\\Downloads\\coding for practice\\Food rec for walmart project\\archive (1)\\1662574418893344.csv')

```

```python
# Function to generate dish recommendations
def recommend_dishes(user_preferences):
    dish_list = df['Name'].tolist()
    prompt_content = f"Based on the following preferences: {user_preferences}, recommend dishes from this list: {dish_list}."
    return get_response(prompt_content) 
```

```python
# Function to get the response from the Groq API
def get_response(prompt_content):
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": prompt_content,
            }
        ],
        model="llama3-70b-8192",
    )
    return chat_completion.choices[0].message.content
```

```python
# Function to suggest dishes with preparation steps
def suggest_dishes_with_steps(user_preferences):
    recommendations = recommend_dishes(user_preferences)
    recommended_dishes = recommendations.split(", ")
    dishes_with_steps = {}
    
    for dish in recommended_dishes:
        steps = df[df['Name'] == dish]['Steps'].values
        if steps:
            dishes_with_steps[dish] = steps[0]
    
    return dishes_with_steps
```

```python
# Streamlit UI
st.title("Dish Recommendation System")
st.write("Enter your preferences to get dish recommendations:")

user_input = st.text_input("Your Preferences", placeholder="e.g., I enjoy veg dishes that contain olive oil and onion.")

if st.button("Get Recommendations"):
    if user_input:
        dish_recommendations_with_steps = suggest_dishes_with_steps(user_input)
        
        if dish_recommendations_with_steps:
            st.write("### Recommended Dishes:")
            for dish, steps in dish_recommendations_with_steps.items():
                st.write(f"**Dish**: {dish}")
                st.write(f"**Preparation Steps**: {steps}")
                st.write("---")
        else:
            st.write("No recommendations found for your preferences.")
    else:
        st.write("Please enter your preferences to get recommendations.")
```
